##### Number of Reviews of Highest Rated iPhones

```python
iphones = highest_rated['product_name'].value_counts()
label = iphones.index
rev_counts = highest_rated["number_of_reviews"]
```

```python
plot = sns.barplot(data=data, x=label, y=rev_counts, palette="Set2")
plt.xlabel("iPhone")
plt.ylabel("Number of Reviews")
plt.title("Number of Reviews of Highest Rated iPhones")
plt.xticks(rotation=90)

for p in plot.patches:
    plot.annotate(format(p.get_height(), '.0f'),
                  (p.get_x() + p.get_width() / 2., p.get_height()),
                  ha = 'center', va = 'center',
                  xytext = (0, -9), textcoords = 'offset points')

plt.show()
```

**APPLE iPhone 8 Plus (Gold, 64 GB)** is also leading in the highest number of reviews on Flipkart among the highest-rated iPhones in India.

##### Relationship between Sale Price and Number of Ratings of iPhones

```python
sns.scatterplot(data = data, x="number_of_ratings", y='sale_price', 
                size = "discount_percentage", sizes=(25,200),
               color="blue", alpha=0.2, edgecolor='black', legend=False)

sns.regplot(data = data, x="number_of_ratings", y='sale_price', scatter=False,
           color='#ea6b6d', ci=None, line_kws={'linewidth': 2})

plt.xlabel("Number of Ratings")
plt.ylabel("Sale Price")
plt.title("Relationship between Sale Price and Number of Ratings of iPhones")


plt.show()
```

There is a **negative** linear relationship between the sale price of iPhones and the number of ratings. It means iPhones with lower sale prices are sold more in India.

##### Relationship between Discount Percentage and Number of Ratings of iPhones

```python
sns.scatterplot(data = data, x="number_of_ratings", y='discount_percentage', 
                size = "sale_price", sizes=(25,200),
               color="blue", alpha=0.2, edgecolor='black', legend=False)

sns.regplot(data = data, x="number_of_ratings", y='discount_percentage', scatter=False,
           color='#ea6b6d', ci=None, line_kws={'linewidth': 2})

plt.xlabel("Number of Ratings")
plt.ylabel("Discount Percentage")
plt.title("Relationship between Discount Percentage and Number of Ratings of iPhones")


plt.show()
```

There is a **positive** linear relationship between the discount percentage on iPhones on Flipkart and the number of ratings. It means iPhones with high discounts are sold more in India.
